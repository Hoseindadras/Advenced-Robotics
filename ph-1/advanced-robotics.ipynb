```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
data_5cm = np.array([1.8, 2.3, 2.8, 2.9, 3.2, 3.4, 3.6, 3.6, 3.7, 3.8, 3.8, 3.9, 3.9, 3.9, 3.9, 4.2, 4.4, 4.4, 4.5, 4.6, 5.4, 5.5, 5.5, 6.5, 6.8, 7, 7.5, 8.4, 9, 13.1])
data_10cm = np.array([6.9, 7.2, 7.4, 7.4, 7.5, 8.2, 8.3, 8.3, 8.3, 8.4, 8.4, 9.5, 9.5, 9.6, 9.6, 9.6, 9.6, 9.6, 9.7, 9.7, 9.8, 9.9, 10, 10.2, 10.4, 11.1, 11.2, 12.4, 12.7, 16.2])
data_15cm = np.array([11.1, 11.2, 12.3, 12.7, 12.8, 13, 13.1, 13.1, 13.1, 13.2, 13.6, 14, 14, 14, 14.1, 14.1, 14.2, 14.4, 14.4, 14.5, 14.7, 15.4, 15.5, 15.5, 15.6, 17.3, 17.3, 17.3, 17.4, 17.5])
```

```python
def calculate_statistics(data):
    mean = np.mean(data)
    variance = np.var(data)
    return mean, variance
```

```python
mean_5cm, var_5cm = calculate_statistics(data_5cm)
mean_10cm, var_10cm = calculate_statistics(data_10cm)
mean_15cm, var_15cm = calculate_statistics(data_15cm)
```

```python
def plot_histogram_and_pdf(data, title):
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=10, density=True, alpha=0.6, color='g', label='Histogram')
    
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, np.mean(data), np.std(data))
    plt.plot(x, p, 'k', linewidth=2, label='PDF')
    
    plt.title(f'Histogram and PDF of {title}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_histogram_and_pdf(data_5cm, '5cm Data')
```

```python
plot_histogram_and_pdf(data_10cm, '10cm Data')
```

```python
plot_histogram_and_pdf(data_15cm, '15cm Data')
```

```python
data_dict = {
    '5cm': data_5cm,
    '10cm': data_10cm,
    '15cm': data_15cm
}
```

```python
df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data_dict.items()]))
df.to_csv('robot_movement_data.csv', index=False)
```

```python
statistics_df = pd.DataFrame({
    'Measurement': ['5cm', '10cm', '15cm'],
    'Mean': [mean_5cm, mean_10cm, mean_15cm],
    'Variance': [var_5cm, var_10cm, var_15cm]
})
```

```python
print("Movement Statistics:")
print(statistics_df)
```

```python
statistics_df.to_csv('movement_statistics.csv', index=False)
```

```python
data_RL = np.array([47, 79, 65, 64, 78, 78, 80, 90, 98, 90])
data_RR = np.array([65, 67, 53, 69, 80, 82, 70, 80, 76, 102])
```

```python
mean_RL, var_RL = np.mean(data_RL), np.var(data_RL)
mean_RR, var_RR = np.mean(data_RR), np.var(data_RR)
```

```python
rotation_statistics_df = pd.DataFrame({
    'Rotation': ['Counterclockwise (RL)', 'Clockwise (RR)'],
    'Mean': [mean_RL, mean_RR],
    'Variance': [var_RL, var_RR]
})
```
